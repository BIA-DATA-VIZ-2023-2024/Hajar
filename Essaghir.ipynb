```python
import pandas as pd
```

```python
# df = pd.read_csv("i_3287.csv")
# df = pd.read_xls("i_3287.xls")
df = pd.read_csv("i_3287.csv")
```

```python
df.head()
```

```python
# Identify the row to be set as the new header (index 0)
new_header = df.iloc[4]

# Skip the rows up to index 3 and set the new header
df = df[5:]
df.columns = new_header

# Resetting the index if needed
df.reset_index(drop=True, inplace=True)
```

```python
print("\nNumber of duplicates:", df.duplicated().sum())
```

```python
print("\nData types:\n", df.dtypes)
```

```python
print("\nNumber of unique values:\n", df.nunique())
```

```python
print("\nDescriptive statistics:\n", df.describe())
```

```python
df = df.dropna()
```

```python
df = df.drop_duplicates()
```

```python
df['2015'] = df['2015'].astype(float)
df['2016'] = df['2016'].astype(float)
```

```python
df.columns = [str(col).replace('4_', '') for col in df.columns]
```

```python
df.to_csv("df_cleaned.csv", index=False)
```
